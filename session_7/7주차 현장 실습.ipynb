```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras

df = pd.read_csv("C:/Users/yoojm/AIStudy/practice/Dataset/diabetes.csv")
```

```python
# 데이터 확인
df
```

```python
# 컬렴명 확인
df.columns
```

```python
# 레이블 갯수 확인 (그래프)
sns.countplot(data = df, x="Outcome")
plt.xlabel("Diabetes or Not")
plt.ylabel("Count")
plt.title("Outcome")
```

```python
# 레이블 갯수 확인 (숫자)
df['Outcome'].value_counts()
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 결측치 제거
df = df.dropna()
df.isnull().sum()
```

```python
# 컬럼별 히스토그램 확인
# 'Insulin'와 'BMI' 칼럼을 선택

figure = plt.figure(figsize=(18,6))
ax1 = plt.subplot(1,2,1)
sns.distplot(df['Insulin'], kde = True, hist = False, ax=ax1)
ax2 = plt.subplot(1,2,2)
sns.distplot(df['BMI'], ax=ax2)
```

```python
# 산점도 확인

import matplotlib.pyplot as plt
import numpy as np

plt.scatter(df['Insulin'], df['BMI'])
plt.xlabel('Insulin')
plt.ylabel('BMI')

plt.show()
```

```python
# pairplot를 사용해 각 상관관계 살펴보기

sns.pairplot(df,
             vars=['Insulin', 'BMI', 'mean perimeter', 'mean area'],
            hue ='label')
plt.show()
```
